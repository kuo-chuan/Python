# 文本斷詞


一篇文本的架構組成分為三個，分別是段落、句子和詞彙，其中段落是由句子群組成，是表達完整的語意片段; 句子是由詞彙群組成，是語言基本單位; 詞彙則是最小語意單位，因此一般分析文本之前，需要透過斷詞演算法將文本轉為詞彙才能被機器所處理。

e.g. "我想去動物園玩" -> ["我", "想", "去", "動物園", "玩"]



:斷詞方式通常分為三種


*   基於字典匹配法: 若文本出現字典中的詞彙則匹配切斷
*   基於規則匹配: 預先定義好斷詞策略，以該策略匹配文本進行斷詞
*   基於模型方法: 由人工標注斷詞資料並訓練斷詞語言模型



```
基於字典或基於規則通常準確率較高，但能涵蓋的範圍非常低
基於模型方法涵蓋範圍廣，但需要大量訓練資料學習
```



## 英文斷詞


````python
!pip install nltk # 英語為主自然語言處理包

# 規則斷詞

# ngram斷詞
'''
e.g. 2-gram

"abcde" -> ["ab", "bc", "cd", "de"]
'''
import nltk
en_df["ngram_seg"] = en_df["text"].apply(lambda x: list(nltk.ngrams(x, 3)))

# 空白斷詞
'''
@sharlynnx aww lets have a twitter party!! -> ["@sharlynnx", "aww", "lets", "have", "a", "twitter party!!"]
'''
from nltk.tokenize import WhitespaceTokenizer
en_df["whitespace_seg"] = en_df["text"].apply(lambda x: WhitespaceTokenizer().tokenize(x))

# 空白+標點符號斷詞
'''
@sharlynnx aww lets have a twitter party!! -> ['@', 'sharlynnx', 'aww', 'lets', 'have', 'a', 'twitter', 'party', '!!']
'''
from nltk.tokenize import WordPunctTokenizer
en_df["wordpunct_seg"] = en_df["text"].apply(lambda x: WordPunctTokenizer().tokenize(x))

# 正規表達式斷詞
'''
e.g.: [A-Z]\w -> [Re]g[Ex]r was created by gskinner.com, and is proudly hosted by [Me]dia [Te]mple.

[A-Z]表示單個大寫字元
\w等價於[A-Za-z0-9_]
'''
from nltk.tokenize import RegexpTokenizer
regexp_tokenizer = RegexpTokenizer('\s+', gaps=True) # \s+表示空白斷詞; gaps表示將regexp當成分隔符號
en_df["regexp_seg"] = en_df["text"].apply(lambda x: regexp_tokenizer.tokenize(x))

# Treebank斷詞器(基於句法結構)
'''
空白斷詞+多種標點符號規則斷詞，使斷詞結果接近正常句構
'''
from nltk.tokenize import TreebankWordTokenizer
en_df["treebank_seg"] = en_df["text"].apply(lambda x: TreebankWordTokenizer().tokenize(x))

# 模型斷詞

# stanford斷詞器
# from nltk.tokenize.stanford import StanfordTokenizer
# !wget https://nlp.stanford.edu/software/stanford-tagger-4.2.0.zip #下載斷詞器模型檔
# !unzip "/content/stanford-tagger-4.2.0.zip" -d "/content"

# stanford_tokenizer =  StanfordTokenizer("/content/stanford-postagger-full-2020-11-17/stanford-postagger.jar")
# en_df["stanford_seg"] = en_df["text"].apply(lambda x: stanford_tokenizer.tokenize(x)) # Stanford tokenizer is slow
````

# 停止詞(Stop Word)

停止詞指的是在文本中可以被忽略而不影響理解的詞彙，通常是用來當作連接詞或
修飾詞用，譬如英文的“the”, "and", "a", 中文的"的", "了", "嗎"。<br><br>
在傳統模型中，會透過移除停止詞去除文本雜訊並減少儲存空間來取得準確率和空間利用率比較平衡的模型，而深度學習模型，則頃向於利用大數據本身優勢自動學習出分辯停止詞和重要詞的能力。

# 字根提取(stemming)

機器擅長處理規則、變異性低的資料，而文本資料是一種語意複雜性高的資料結構，不同用詞能夠組合出相同意義句子，譬如在英文文法中詞彙有時態變化，像是play, played, playing，對於人類來講是相同含意，但對於機器來說卻是三種全新字詞，這會增加模型在資料學習上的難度。

字根提取技術可以提取字詞變化中共同部分，譬如**「play, played, playing」可以提出"play"**，**「plan, planed, planning」可以提出"plan"**部分，但提取出的共同部分不一定為合法詞彙，像是「football」可以提取出"footbal"，但在字典中並無"footbal"詞彙。

````python
# Porter stemmer
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
en_df["porter_stem"] = en_df["wordpunct_seg"].apply(lambda x: [porter_stemmer.stem(word) for word in x])

# Snowball stemmer
from nltk.stem.snowball import SnowballStemmer
snowball_stemmer = SnowballStemmer("english")
en_df["snowball_stem"] = en_df["wordpunct_seg"].apply(lambda x: [snowball_stemmer.stem(word) for word in x])

# Lancaster
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
en_df["lancaster_stem"] = en_df["wordpunct_seg"].apply(lambda x: [lancaster_stemmer.stem(word) for word in x])
````

# 字詞還原(Lemmatisation)

字詞還原指將英文詞彙還原為詞形，詞形表示詞彙的原始形式，譬如cats還原為cat、came還原為come，字詞還原跟字根提取目的相同都是降低用詞複雜性，不同的是字詞還原的詞形是合法詞彙存在於在字典中。

詞形因為有實際意義存在，在詞彙操作上比字根提取後的結果會更有意義，若有精確建模需求，則使用字詞還原會更合適。

````python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

en_df["lemmatisation"] = en_df["wordpunct_seg"].apply(lambda words: [lemmatizer.lemmatize(word) for word in words])
````

## 詞頻統計

````python
# hotel正向評論詞頻統計
word_count = {}
for sent in pos_hotel_df["精確模式"]:
  for word in sent:
    if word not in word_count:
      word_count[word] = 0
    word_count[word] += 1 

# 排序詞頻
sort_word_count = sorted(word_count.items(), key=lambda x:x[1], reverse=True)
display(sort_word_count)
# # 視覺化
ax = pd.DataFrame(sort_word_count)[1].plot(figsize=(15,5))
ax.set_xlim(0,200)
# # 少量詞彙構成生活中的用語

def word_count_analysis(df, col):
  word_count = {}
  for sent in df[col]:
    for word in sent:
      if word not in word_count:
        word_count[word] = 0
      word_count[word] += 1 

  # 排序詞頻
  sort_word_count = sorted(word_count.items(), key=lambda x:x[1], reverse=True)
  display(sort_word_count)
  # 視覺化
  ax = pd.DataFrame(sort_word_count)[1].plot(figsize=(15,5))
  ax.set_xlim(0,200)
  
  return word_count
````

````python
# hotel負向評論詞頻統計
neg_hotel_word_count = word_count_analysis(neg_hotel_df, "精確模式")
````

````python
# twitter留言詞頻統計
en_word_count = word_count_analysis(en_df, "wordpunct_seg")
````

# 文本表達式

文字資料是人類溝通所使用的資料格式，是只有人類才能讀懂的，像是報章雜誌、新聞文本等，而人類可讀資料格式不代表機器可讀，因為機器只能讀懂數字。
在NLP領域裡語言模型就代表機器，而模型可以存取的格式是數學上被稱作『向量/矩陣』的格式，在Python世界中通常會編碼成『Vector/DataFrame』格式來存放。


## 獨熱編碼(one-hot)

獨熱編碼是一種二進制向量，且在任意維度中只允許其中一個維度為1，其餘為0，在數位電路中常用來表達不同的電路狀態，在文本中可以用來代表不同詞彙，操作上可以分成下列兩個步驟:

1.   收集用來表示狀態的詞彙集合
2.   文本斷詞

## 獨熱編碼改進


獨熱編碼除了會造成大量不必要的0位元資訊，稀疏的1位元在字詞間上不容易找到關聯性，透過適當的空間壓縮可以提升空間效用同時增強詞彙關聯性，實做重點上有以下兩個:


1.   一個句子表示為一筆資料
2.   句子中出現的詞彙皆編碼為有效位元

第一點是縮減列向量(row)空間，重新定義句子為一筆資料並存成一個列空間

第二點是增加有效位元量，原獨熱編碼是一列向量只有一有效位元，造成稀疏程度高，壓縮資料後可以同時存在多個有效位元。

<img src="https://drive.google.com/uc?export=view&id=1XARRmTf6cGIbW8k_Xr4y1c8RszoXYQAh"  width="800" height="350" align="center" />

````python
# twitter comment one hot encoding
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer=lambda x:x, preprocessor=lambda x:x, binary=True)
en_one_hot = vectorizer.fit_transform(en_df["wordpunct_seg"])
display(vectorizer.get_feature_names_out())
print(en_one_hot.toarray())

def one_hot_encoding(sents):
  vectorizer = CountVectorizer(tokenizer=lambda x:x, preprocessor=lambda x:x, binary=True)
  one_hot = vectorizer.fit_transform(sents)
  display(vectorizer.get_feature_names_out())
  print(one_hot.toarray())
  return vectorizer, one_hot
````

## TF-IDF編碼
TF-IDF是一種結合文字加權技術的資料特徵擷取方法，以統計方式評估詞彙在不同文本中的重要性，背後假設詞彙在不同文本出現的頻率變化會影響詞彙重要性程度，是傳統文本分析應用常用的文本表示法，像是資料檢索、文本分類、關鍵字擷取、文本相似度計算等。


**Term Frequency(TF)編碼**<br><br>

一般在文本中經常出現的詞彙會被認為是重要的詞，當出現次數越高表示越重要，TF
就是用來衡量詞彙在文本中出現頻率的一種指標，而TF編碼就是將詞頻資訊編碼到向量中。

<p align="center">
<img src="https://drive.google.com/uc?export=view&id=1HLpQ7MQUWkbO4M3ZEdeIC1N6bKk4WCTb"  width="800" height="200" align="center" />

````python
# twitter comment Tf encoding
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(tokenizer=lambda x:x, preprocessor=lambda x:x, norm=None, use_idf=False)
en_tfidf = tfidf_vectorizer.fit_transform(en_df["wordpunct_seg"])
display(tfidf_vectorizer.get_feature_names_out())
print(en_tfidf.toarray())
print(tfidf_vectorizer.inverse_transform(en_tfidf[0]))

def tf_encoding(sents):
  tfidf_vectorizer = TfidfVectorizer(tokenizer=lambda x:x, preprocessor=lambda x:x, norm=None, use_idf=False)
  tf = tfidf_vectorizer.fit_transform(sents)
  display(tfidf_vectorizer.get_feature_names_out())
  print(tf.toarray())
  print(tfidf_vectorizer.inverse_transform(tf[0]))
  return tfidf_vectorizer, tf
````

````python
# positive hotel comment Tf encoding
display(tf_encoding(pos_hotel_df["精確模式"]))

# negative hotel comment Tf encoding
display(tf_encoding(neg_hotel_df["精確模式"]))
````

**Inverse Document Frequency(IDF)編碼**

實際統計詞頻後會發現，詞頻高的有些是不具有重要意義的詞，譬如英文的“the”、“and”、”an”，中文的"了"、"的"、"嗎"之類的詞彙，這些詞常經常是連接詞、語助詞，因此會需要其他方法過濾這些詞，像是Inverse Document Frequency。

IDF是一種評估詞彙在跨文本中重要性的方法，當多個文本出現同樣的詞彙，代表該詞彙在多數文本中不具有區別性意義，譬如"the"、"and"、"了"、"的"之類的詞彙，其重要性分數應該降低。

## TF-IDF
同時結合TF和IDF特徵編碼方式，實做上會取TF值乘上IDF值做為文本編碼值。

<p align="center">
<img src="https://drive.google.com/uc?export=view&id=1cODyOR5kDGptoZ0o43sm7m5TloHfk-6m"  width="800" height="400" align="center" />

````python
def tfidf_encoding(sents):
  tfidf_vectorizer = TfidfVectorizer(tokenizer=lambda x:x, preprocessor=lambda x:x, norm=None, use_idf=True)
  tfidf = tfidf_vectorizer.fit_transform(sents)
  display(tfidf_vectorizer.get_feature_names_out())
  print(tfidf.toarray())
  print(tfidf_vectorizer.inverse_transform(tfidf[0]))
  return tfidf_vectorizer, tfidf
````

````python
# twitter comment Tf-IDF encoding
display(tfidf_encoding(en_df["wordpunct_seg"]))

# positive hotel comment Tf-IDF encoding
display(tfidf_encoding(pos_hotel_df["精確模式"]))

# negative hotel comment Tf-IDF encoding
display(tfidf_encoding(neg_hotel_df["精確模式"]))
````

## word embedding(詞嵌入)

詞嵌入是一種將高維度映射到低維度的文本表示方法，結合特徵學習和自然語言模型技術，目的是想藉由資料驅動方式自動從文本中學習詞彙向量每個維度的自然含義。

詞嵌入向量有以下3種目標:


1.   詞向量維度低(<< 詞袋大小)
2.   詞向量保留詞和詞之間的語意關係
3.   **相似語意詞向量距離接近**

以詞袋做為表示法容易受到大詞袋而有向量稀疏問題，體現在向量間出現正交情況，因此後續NLP發展目標轉為尋找低維度空間表示法，維度降維讓詞彙群能夠在空間中更緊密貼近而不發散，衡量向量物理涵義才有意義，譬如距離、角度。
